```python
import numpy as np
import scipy.stats
```

```python
def mean_confidence_interval(data,confidence=.95):
    a=1.0*np.array(data)
    n=len(a)
    m,se=np.mean(a), scipy.stats.sem(a)
    h=se*scipy.stats.t.ppf((1+confidence)/2., n-10)
    return m,m-h,m+h
```

```python
mean_confidence_interval(data_total["tripduration"])
```
